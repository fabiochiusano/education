# Calculate complexity for each article

```python
import nbimporter
from constants import Constants
from preprocessed_data_reader import ReaderPreprocessedData

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
dataset = ReaderPreprocessedData.read_data("../preprocessed")
```

```python
for sample in dataset:
    sample["complexity"] = np.log(sample["tfidf"]["logtfidf"] + 1.000001).mean()
```

```python
dataset = sorted(dataset, key=lambda s:s["complexity"])
```

```python
# simple articles
for sample in dataset[:5]:
    print("{0} --> {1}".format(sample["complexity"], sample["title"]))
    plt.hist(np.log(sample["tfidf"]["logtfidf"] + 1.000001))
    plt.show()
    print(sample["url"])
    print(sample["content"])
    print("---------------------------")
```

```python
# complex articles
for sample in dataset[::-1][:5]:
    print("{0} --> {1}".format(sample["complexity"], sample["title"]))
    plt.hist(np.log(sample["tfidf"]["logtfidf"] + 1.000001))
    plt.show()
    print(sample["content"])
    print("---------------------------")
```
